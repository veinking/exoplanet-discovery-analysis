# Exoplanet Analysis

Goal:
Analyze confirmed exoplanets to understand size, stellar energy, and detection confidence.

Steps:
- Filter confirmed planets
- Clean relevant features
- Create size categories
- Analyze distributions
- Export for Tableau

Import block


```python
import pandas as pd
import numpy as np
from google.colab import files
```

Collab Block

```python
from google.colab import drive
path = '/content/drive/MyDrive/exoplanets.csv'

df = pd.read_csv(path)
```

Column Exploration

```python

df.columns.tolist()
```

# filter confirmed discoveries only and keep columns relevant to question


```python
df_raw = df.copy()

df_confirmed = df_raw[df_raw['koi_disposition'] == 'CONFIRMED']

df_confirmed = df_confirmed[['kepoi_name','koi_prad','koi_insol','koi_score']].dropna()

df_confirmed = df_confirmed.sort_values(by='koi_score', ascending=False)
df_confirmed.head()
```

Size Categories

```python
df_confirmed['size_category'] = pd.cut(
    df_confirmed['koi_prad'],
    bins=[0,1.5,3,6,20]
    ,labels=['earth-like','super-earth-like','neptune-like','Gas-Giant']

)
df_confirmed.head()
```

Distributions

```python
size_dist = df_confirmed['size_category'].value_counts().reset_index()
size_dist.columns = ['size_category','count']
size_dist
```

Stellar Energy

```python
energy_bins = pd.cut(
    df_confirmed['koi_insol'],
    bins=[0,10,100,500,2000],
    labels=['Low','Moderate','High','Extreme']
)
energy_dist = energy_bins.value_counts().reset_index()
energy_dist.columns = ['energy_level', 'count']
energy_dist
```

Confidence by size

```python
confidence_by_size = df_confirmed.groupby('size_category')['koi_score'].mean().reset_index()
confidence_by_size
```

Export and download clean csv

```python
pd.DataFrame.to_csv(confidence_by_size, 'confidence_by_size.csv')
pd.DataFrame.to_csv(df_confirmed, 'confirmed_exoplanets.csv')
pd.DataFrame.to_csv(energy_dist, 'energy_dist.csv')
pd.DataFrame.to_csv(size_dist, 'size_dist.csv')
```
